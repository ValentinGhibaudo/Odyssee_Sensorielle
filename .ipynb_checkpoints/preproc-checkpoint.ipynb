```python
save = True
```

## TOOLS

```python
def preprocessing(sig, dtype):
    if dtype == 'ECG':
        clean = nk.ecg_clean(sig)
    elif dtype == 'EDA':
        clean = nk.eda_clean(sig)
    elif dtype == 'PPG':
        clean = nk.ppg_clean(sig)
    elif dtype == 'EEGL' or dtype == 'EEGR':
        # clean = filter_sig(sig=sig,fs=1000, low=0.1, high=10 , order=4, btype = '40', show = False)
        clean = mne.filter.notch_filter(sig, 1000, freqs=np.arange(50,201,50), verbose = False) # notch filter to remove line_noise
        clean = mne.filter.filter_data(clean, 1000, l_freq = 0.1, h_freq = None, verbose = False) # remove slow drift with high pass 0.1
                
    return clean
```

```python
def time_duration_by_participant():
    rows = []
    for participant in participants:
        duration = float(da_all.sel(participant = participant).dropna(dim='time').coords['time'][-1].values)
        rows.append([participant, duration])
    df = pd.DataFrame(rows, columns = ['participant','duration']).set_index('participant')
    return df.round(2)
time_duration_by_participant()
```

```python
def get_rooms_time_caracteristics(participants, rooms, save=False):
    
    durations = time_duration_by_participant()
    
    def get_sec(time_str):
        """Get seconds from time."""
        h, m, s = time_str.split(':')
        return int(h) * 3600 + int(m) * 60 + int(s)

    path = '../data/timings_hours.xlsx'
    df = pd.read_excel(path, index_col = [0,1])
    
    secs_between_exit_and_stop_recording = 60
    
    rows = []
    for participant in participants:
        duration_participant = int(durations.loc[participant, 'duration'])
        # print(duration_participant)
        end_of_recording = get_sec(str(df.loc[(participant,'End of the world'),'stop'])) + secs_between_exit_and_stop_recording
        
        for room in rooms:     
            time_start = df.loc[(participant,room),'start']
            secs_start = duration_participant - (end_of_recording - get_sec(str(time_start)))
            time_stop = df.loc[(participant,room),'stop']
            secs_stop = duration_participant - (end_of_recording - get_sec(str(time_stop)))
            duration_in_room = secs_stop - secs_start 
            row = [participant, room, secs_start , secs_stop, duration_in_room]
            rows.append(row)
    df_return = pd.DataFrame(rows, columns = ['participant', 'room', 'start','stop','duration in room']).set_index(['participant','room'])
    if save:
        df_return.to_excel('../data/timings.xlsx')
    return df_return

def get_sel_timings(participants, rooms):
    
    durations = time_duration_by_participant()
    
    def get_sec(time_str):
        """Get seconds from time."""
        h, m, s = time_str.split(':')
        return int(h) * 3600 + int(m) * 60 + int(s)

    path = '../data/timings_hours.xlsx'
    df = pd.read_excel(path, index_col = [0,1])
    
    secs_between_exit_and_stop_recording = 60
    
    rows = []
    for participant in participants:
        duration_participant = int(durations.loc[participant, 'duration'])
        # print(duration_participant)
        end_of_recording = get_sec(str(df.loc[(participant,'End of the world'),'stop'])) + secs_between_exit_and_stop_recording
        
        for room in rooms:     
            time_start = df.loc[(participant,room),'start_sel']
            secs_start = duration_participant - (end_of_recording - get_sec(str(time_start)))

            row = [participant, room, secs_start]
            rows.append(row)
            
    df_return = pd.DataFrame(rows, columns = ['participant', 'room', 'start']).set_index(['participant','room'])

    return df_return
```

```python
das = []
for participant in participants:
    da = xr.load_dataarray(f'../data_in_da/da_{participant}.nc')
    das.append(da)
da_all = xr.concat(das, dim = 'participant')
da_all = da_all.assign_coords({'participant':participants})
```

```python
da_cleans = []
for participant in participants:
    da_clean = None
    print(participant)
    for dtype in dtypes:
        print(dtype)
        sig = da_all.loc[participant, dtype, :].dropna(dim='time').values
        # print(sig.size)
        clean = preprocessing(sig, dtype)
        if da_clean is None:
            da_clean = init_da({'dtype':dtypes, 'time':time_vector(sig, 1000)})
        da_clean.loc[dtype, :] = clean
    da_cleans.append(da_clean)
da_cleans = xr.concat(da_cleans, dim = 'participant')
da_cleans = da_cleans.assign_coords({'participant':participants})
da_both = xr.concat([da_all, da_cleans], dim = 'cleaning')
da_both = da_both.assign_coords({'cleaning':['raw','clean']})
```

```python
da_both
```

```python
start_timings = get_sel_timings(participants,rooms)
time = np.arange(0, data_sel_duration, 1/srate)

da_preprocessed = None

for participant in participants:
    for room in rooms:
        start = start_timings.loc[(participant, room),'start']
        stop = start + data_sel_duration
        data_room = da_both.loc['clean', participant,:,start:stop].values[:,:-1]

        
        if da_preprocessed is None:
            da_preprocessed = init_da({'participant':participants, 'room':rooms, 'dtype':dtypes, 'time':time})
            
        da_preprocessed.loc[participant, room, :, :] = data_room
```

```python
if save:
    da_preprocessed.to_netcdf('../data_preprocessed/da_preprocessed.nc')
```
